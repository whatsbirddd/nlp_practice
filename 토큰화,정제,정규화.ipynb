```
import nltk
nltk.download('punkt')
```

```
#단어 토큰화 : 영어 코퍼스 1. word_tokenize
from nltk.tokenize import word_tokenize
print(word_tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
```

```
#단어 토큰화 : 영어 코퍼스 2. WordPunctTokenizer
from nltk.tokenize import WordPunctTokenizer
print(WordPunctTokenizer().tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
```

```
#단어 토큰화: 영어 코퍼스 3. text_to_word_sequence (keras)
from tensorflow.keras.preprocessing.text import text_to_word_sequence
print(text_to_word_sequence("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
```

```
#문장 토큰화
from nltk.tokenize import sent_tokenize
text = "His barber kept his word. But keeping such a huge secret to himself was driving him crazy. Finally, the barber went up a mountain and almost to the edge of a cliff. He dug a hole in the midst of some reeds. He looked about, to make sure no one was near."
print(sent_tokenize(text))
```

```
#중간에 마침표가 여러개인경우 -> ph.D.를 단어로 인식함
from nltk.tokenize import sent_tokenize
text="I am actively looking for Ph.D. students. and you are a Ph.D student."
print(sent_tokenize(text))
```

```
#konlpy 형태소 분석기 : okt(open korea text),mecab,komoran,hannanum,kkma
from konlpy.tag import Okt
okt = Okt()
print(okt.morphs("열심히 코딩한 당신, 연휴에는 여행을 가봐요")) #형태소 추춘
```

```
print(okt.pos("열심히 코딩한 당신, 연휴에는 여행을 가봐요")) #태깅
```

```
print(okt.nouns("열심히 코딩한 당신, 연휴에는 여행을 가봐요")) #명사추출
```

```
from konlpy.tag import Kkma
kkma=Kkma()
print(kkma.morphs("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
```

```
print(kkma.nouns("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
```

# cleaning and Normalization
1. 규칙에 기반한 표기가 다른 단어들의 통합
- 어간 추출(stemming), 표제어 추출(lemmatization)
2. 대,소문자 통합
- Automobile을 검색하면 automobile도 검색가능
- 하지만 US와 us는 구별되어야 하므로 주의
- 회사이름, 사람이름 (General Motors, Bush)등은 대문자로 유지하는 것이 좋음
- 해결방법: 문장의 맨 앞에 나오는 단어의 대문자만 소문자로 변환하고 나머지는 대문자인 상태로 두는 것
3. 불필요한 단어(noise data)의 제거 : 불용어 제거 part에서 다룰 것
- removing rare words
  스팸메일을 분류하는데 스팸메일에 자주나타나는 단어가 아니라 1번 나타날까말까한 단어를 두는 것은
